```python
import pandas as pd
import folium
```

```python
folium_map = folium.Map(location=[35.690921, 139.700258], zoom_start=15)
folium.Marker(location=[35.690921, 139.700258],
    popup='Shinjuku Station',
    icon=folium.Icon(color='blue', icon='info-sign')).add_to(folium_map)
folium_map
```

```python
folium_map = folium.Map(location=[35.690921, 139.700258],zoom_start=15)

folium.CircleMarker(
    location=[35.690921, 139.700258],
    radius=60,
    color='#ff0000',
    fill_color='#0000ff'
).add_to(folium_map)

folium_map
```

```python
df = pd.DataFrame({
    'station': ['渋谷駅','新宿駅', '上野駅', '東京駅', '所沢駅'],
    'latitude': [35.658034, 35.690921, 35.713768, 35.681236, 35.795969],
    'longtude': [139.701636, 139.700258, 139.777254, 139.767125, 139.473947],
})

folium_map = folium.Map(location=[35.690921, 139.700258], zoom_start=12)

for i, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longtude']],
        popup=row['station'],
        icon=folium.Icon(color='red')
    ).add_to(folium_map)

folium_map
```
